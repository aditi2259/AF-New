```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
import os
```

```python
ecg_directory = '/content/training_data/training2017'
ecg_mat_file = os.path.join(ecg_directory, 'A03852.mat')  # Adjust this file name as necessary
ecg_data = loadmat(ecg_mat_file)
ecg_signal = ecg_data['val'][0]
```

```python
sampling_rate = 500
time_interval = 1
samples_per_segment = sampling_rate * time_interval
```

```python
num_segments = len(ecg_signal) // samples_per_segment  #  number of segments calculation
segmented_signals = np.array_split(ecg_signal, num_segments)
```

```python
plt.figure(figsize=(12, 8))
for i in range(min(5, num_segments)):
    plt.subplot(5, 1, i+1)  # subplots
    plt.plot(segmented_signals[i], color='b')
    plt.title(f'Segment {i+1}')
    plt.xlabel('Sample Index')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
